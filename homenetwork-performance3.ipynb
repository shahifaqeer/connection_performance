```python
from __future__ import division
import pandas as pd
from datetime import datetime
```

###Util functions
- load data
- plot bandwidth
- plot number of packets/bytes
- plot cdf

```python
#load data
def loadData(ASlog):
    """load iperf log csv and returns pandas dataframe

    Dataframe fields: timestamp, srcip, sport, dstip, dport, ignore,
    time, bytes, bandwidth
    Parameter: log file (location/name)"""
    
    df = pd.read_csv(ASlog, skipinitialspace=True, names=['timestamp','srcip','sport', 'dstip', 'dport', 'ignore',
    'time', 'bytes', 'bandwidth'])
    df["datetime"] = df.timestamp.apply(lambda x: datetime(int(str(x)[:4]), int(str(x)[4:6]), int(str(x)[6:8]), int(str(x)[8:10]), int(str(x)[10:12]), int(str(x)[12:])))
    #df = df.set_index('datetime')
    return df
```

```python
#Plot bandwidth
def plotBW(df, title):
    """plot bandwidth in bps from consolidated bandwidth dataframe
    
    dataframe contains 5 colums: AS, AR, RS, A_ARRS, R_ARRS

    Parameters: dataframe, title"""
    figure(figsize=(10,6))
    p = df.AS.plot(marker='o', lw=1.5, color='k')
    p = df.AR.plot(marker='v', ls='--', lw=3, color='r')
    p = df.RS.plot(marker='^', ls='--', lw=3, color='g')
    p = df.A_ARRS.plot(marker='d', ls='-.', lw=3, color='b')
    p = df.R_ARRS.plot(marker='s', ls='-.', lw=3, color='b')
    
    p.set_ylabel("TCP Bandwidth (Bps)")
    p.set_title(title)
    legend()
    return
```

```python
#Plot bandwidth
def plotCDF(X):
    """plot cdf from series X"""
    X.sort()
    Y = X/X.sum()
    CY = Y.cumsum()
    plot(X,CY)
    return

def plotDataFrameCDF(df):
    """plot cdf bandwidth/base_bandwidth"""
    base_column = "AS"
    for column_name in ["AR", "RS", "A_ARRS", "R_ARRS"]:
        X = (df[column_name]/df[base_column]).copy()
        X.sort()
        Y = X/X.sum()
        CY = Y.cumsum()
        plot(X,CY,label=column_name+"/"+base_column)
    return
```

```python
def loadScenario(num):
    #Load all 5
    AS = loadData('testlogs/bandwidth_tests'+str(num)+'/AStest'+str(num)+'.log')
    AR = loadData('testlogs/bandwidth_tests'+str(num)+'/ARtest'+str(num)+'.log')
    RS = loadData('testlogs/bandwidth_tests'+str(num)+'/RStest'+str(num)+'.log')
    A_ARRS = loadData('testlogs/bandwidth_tests'+str(num)+'/A_ARRStest'+str(num)+'.log')
    R_ARRS = loadData('testlogs/bandwidth_tests'+str(num)+'/R_ARRStest'+str(num)+'.log')
    return pd.DataFrame({'AS': AS.bandwidth, 'AR': AR.bandwidth, 'RS': RS.bandwidth,
                'A_ARRS': A_ARRS.bandwidth , 'R_ARRS': R_ARRS.bandwidth})
```

```python
#plot number of packets and/or bytes
def plotBytes(df, title):
    figure(figsize=(10,6))
    p1=df["bytes"].plot(label='iperf-bytes')
    df["router-bytes"].plot(color='g', axes=p1, label='router-bytes')
    p1.set_ylabel("Bytes transferred")
    p1.set_title(title)
    show()
    return
```
