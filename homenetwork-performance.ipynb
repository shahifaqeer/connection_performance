```python
from __future__ import division
import pandas as pd
from datetime import datetime
```

##Plotting Functions

```python
def prelimDataPlots(ASlog):
    df = pd.read_csv(ASlog, skipinitialspace=True, names=['timestamp','srcip','sport', 'dstip', 'dport', 'ignore',
    'time', 'bytes', 'bandwidth', 'router-bytes', 'router-packets'])
    df["datetime"] = df.timestamp.apply(lambda x: datetime(int(str(x)[:4]), int(str(x)[4:6]), int(str(x)[6:8]), int(str(x)[8:10]), int(str(x)[10:12]), int(str(x)[12:])))
    df = df.set_index('datetime')
    return df
```

```python
def plotBW(df, title):
    figure(figsize=(10,6))
    p = df.bandwidth.plot()
    p.set_ylabel("TCP Bandwidth (Bps)")
    p.set_title(title)
    show()
    return

def plotBytes(df, title):
    figure(figsize=(10,6))
    p1=df["bytes"].plot(label='iperf-bytes')
    df["router-bytes"].plot(color='g', axes=p1, label='router-bytes')
    p1.set_ylabel("Bytes transferred")
    p1.set_title(title)
    show()
    return
    
```

## WIRED

```python
ASWired = prelimDataPlots('testlogs/AS-wired.log')
plotBW(ASWired, "Wired AS test")
plotBytes(ASWired,"Wired AS test")
```
