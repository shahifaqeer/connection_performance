```python
from __future__ import division
import pandas as pd
from datetime import datetime
```

###Util functions
- load data
- plot bandwidth
- plot number of packets/bytes
- plot cdf

```python
#load data
def loadData(ASlog):
    """load iperf log csv and returns pandas dataframe

    Dataframe fields: timestamp, srcip, sport, dstip, dport, ignore,
    time, bytes, bandwidth
    Parameter: log file (location/name)"""
    
    df = pd.read_csv(ASlog, skipinitialspace=True, names=['timestamp','srcip','sport', 'dstip', 'dport', 'ignore',
    'time', 'bytes', 'bandwidth'])
    df["datetime"] = df.timestamp.apply(lambda x: datetime(int(str(x)[:4]), int(str(x)[4:6]), int(str(x)[6:8]), int(str(x)[8:10]), int(str(x)[10:12]), int(str(x)[12:])))
    #df = df.set_index('datetime')
    return df
```

```python
#Plot bandwidth
def plotBW(df, title):
    """plot bandwidth in bps from consolidated bandwidth dataframe
    
    dataframe contains 5 colums: AS, AR, RS, A_ARRS, R_ARRS

    Parameters: dataframe, title"""
    figure(figsize=(10,6))
    p = df.AS.plot(marker='o', lw=1.5, color='k')
    p = df.AR.plot(marker='v', ls='--', lw=3, color='r')
    p = df.RS.plot(marker='^', ls='--', lw=3, color='g')
    p = df.A_ARRS.plot(marker='d', ls='-.', lw=3, color='b')
    p = df.R_ARRS.plot(marker='s', ls='-.', lw=3, color='b')
    
    p.set_ylabel("TCP Bandwidth (Bps)")
    p.set_title(title)
    legend()
    return
```

```python
#Plot bandwidth
def plotBW(df, title):
    """plot bandwidth in bps from consolidated bandwidth dataframe
    
    dataframe contains 5 colums: AS, AR, RS, A_ARRS, R_ARRS

    Parameters: dataframe, title"""
    figure(figsize=(10,6))
    p = df.AS.plot(marker='o', lw=1.5, color='k')
    p = df.AR.plot(marker='v', ls='--', lw=3, color='r')
    p = df.RS.plot(marker='^', ls='--', lw=3, color='g')
    p = df.A_ARRS.plot(marker='d', ls='-.', lw=3, color='b')
    p = df.R_ARRS.plot(marker='s', ls='-.', lw=3, color='b')
    
    p.set_ylabel("TCP Bandwidth (Bps)")
    p.set_title(title)
    legend()
    return
```

```python
#plot number of packets and/or bytes
def plotBytes(df, title):
    figure(figsize=(10,6))
    p1=df["bytes"].plot(label='iperf-bytes')
    df["router-bytes"].plot(color='g', axes=p1, label='router-bytes')
    p1.set_ylabel("Bytes transferred")
    p1.set_title(title)
    show()
    return
```

####(a) Good Channel
- logs in 'testlogs/bandwidth_tests3/'
- 5 logs to be loaded
 - AS test: iperf client log at A
 - AR test: iperf client log at A
 - RS test: iperf client log at R
 - ARRS test (A): iperf client log at A for simultaneous AR + RS test
 - ARRS test (R): iperf client log at R for simultaneous AR + RS test
- load logs and consolidate all bandwidth data

```python
#example
AS_good = loadData('testlogs/bandwidth_tests3/AStest3.log')
AS_good.head()
```

```python
#Load all 5
AS_good = loadData('testlogs/bandwidth_tests3/AStest3.log')
AR_good = loadData('testlogs/bandwidth_tests3/ARtest3.log')
RS_good = loadData('testlogs/bandwidth_tests3/RStest3.log')
A_ARRS_good = loadData('testlogs/bandwidth_tests3/A_ARRStest3.log')
R_ARRS_good = loadData('testlogs/bandwidth_tests3/R_ARRStest3.log')
```

```python
#consolidate bandwidth data
scenario1_good_bandwidth = pd.DataFrame({'AS': AS_good.bandwidth, 'AR': AR_good.bandwidth, 'RS': RS_good.bandwidth,
                'A_ARRS': A_ARRS_good.bandwidth , 'R_ARRS': R_ARRS_good.bandwidth})
```

```python
#Plot AStest Bandwidth
plotBW(scenario1_good_bandwidth, "Scenario 1: wireless (A) + wired (B) in good channel")
```
